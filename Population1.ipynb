```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/Users/pratiksha/Documents/Internships/Prodigy Infotech/Task 1 Dataset - Population/Population - Worldbank.csv')
df = df.drop(columns = ['Country Code','Indicator Code','Indicator Name'], axis = 1)
print(df)
```

```python
df = df.dropna()
print(df)
```

```python
years = [str(year) for year in range(1960, 2023, 5)]
india_data = df[df['Country Name'] == 'India']
india_population = india_data[years].squeeze()
plt.figure(figsize=(10, 7))
plt.bar(india_population.index, india_population, color='lightblue', edgecolor = 'black')
plt.title('Population Changes in India (1960-2020)')
plt.xlabel('Year')
plt.ylabel('Population (in 100 crores)')
plt.show()
```

```python
years = [str(year) for year in range(1960, 2023, 5)]
pak_data = df[df['Country Name'] == 'Pakistan']
pak_population = pak_data[years].squeeze()
plt.figure(figsize=(10, 7))
plt.bar(pak_population.index, pak_population, color='orange', edgecolor='black', label='Pakistan Population')
plt.title('Population Changes in Pakistan (1960-2020)')
plt.xlabel('Year')
plt.ylabel('Population (in 10 crores)')
plt.legend()
plt.show()
```
